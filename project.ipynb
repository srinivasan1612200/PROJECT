```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as pt
```

```python
%matplotlib inline
```

```python
ram=pd.read_csv("solarprediction.csv")
```

```python
ram
```

```python
newdataset=ram.drop(['UNIXTime','Data','Time','Radiation','Pressure','WindDirection(Degrees)','Speed','TimeSunRise','TimeSunSet'],axis=1)
```

```python
x=newdataset.iloc[:,0].values.reshape(-1,1)
y=newdataset.iloc[:,1].values.reshape(-1,1)
```

```python
x
```

```python
y
```

```python
pt.scatter(x,y)
pt.xlabel('Temperature')
pt.ylabel('Humidity')
pt.title('linear reg')
pt.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(x,y,train_size=0.8,random_state=0)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
Y_train
```

```python
Y_test
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train,Y_train)
```

```python
y_predict=lm.predict(X_test)
```

```python
y_predict
```

```python
nan=pd.DataFrame(X_test,columns=['Temperature'])
```

```python
nan
```

```python
nan['PRDATA']=y_predict
```

```python
nan
```

```python
pt.scatter(x,y,color='red')
pt.plot(X_test,y_predict,color='blue')
pt.xlabel('Temperature')
pt.ylabel('Humidity')
pt.title('linear reg')
pt.show()
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
mean_squared_error(Y_test,y_predict)
```

```python
mean_absolute_error(Y_test,y_predict)
```

```python
r2_score(Y_test,y_predict)
```

```python
lm.score(X_test,Y_test)
```
